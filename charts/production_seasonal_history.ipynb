```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
```

```python
month_map = {
    1: 'Jan.',
    2: 'Feb.',
    3: 'Mar.',
    4: 'Apr.',
    5: 'May.',
    6: 'Jun.',
    7: 'Jul.',
    8: 'Aug.',
    9: 'Sept.',
    10: 'Oct.',
    11: 'Nov.',
    12: 'Dec.'
}
```

```python
df = pd.read_csv('../data/electricity_monthly.csv')
df.tail()
df['month'] = df['month'].apply(lambda x: month_map.get(x))
```

```python
df = df.loc[(df['year_month'] >= '2016-04') & (df['year_month'] < '2019-10')]
df.tail()
```

```python
production_therm_and_renewable = {
    'name': 'Therm. and Renewables',
    'type': 'bar',
    'x': [df['year_season'], df['month']],
    'y': df['production_therm_and_renewable'],
    'marker': {
        'color': '#c75b4f'
    } 
}

production_nuclear = {
    'name': 'Nuclear',
    'type': 'bar',
    'x': [df['year_season'], df['month']],
    'y': df['production_nuclear'],
    'marker': {
        'color': '#c7974f'
    } 
}


production_hydro = {
    'name': 'Hydro',
    'type': 'bar',
    'x': [df['year_season'], df['month']],
    'y': df['production_hydro'] - df['production_hydro_losses'],
    'marker': {
        'color': '#4f99c7'
    } 
}

consumption = {
    'name': 'Consumption',
    'x': [df['year_season'], df['month']],
    'y': df['consumption_country'],
    'mode': 'lines',
    'line_color': '#000000',
    'line': {
        'width': 3
    },
    'showlegend': False,py
}

consumption_point = {
    'name': 'Consumption Label',
    'x': [df['year_season'].iloc[-1], df['month'].iloc[-1]],
    'y': [df.consumption_country.iloc[-1]],
    'mode': 'markers',
    'marker': {
        'color': '#000000'
    },
    'showlegend': False,
    'hoverinfo': 'skip'
}



layout = {
    'barmode': 'stack',
    'title': 'Production Seasonal History',
    'plot_bgcolor': '#ffffff',
    'showlegend': True,
    'legend': {
        'orientation': 'h',
        'y': -0.4,
        
    },
    'margin': {
        'r': 250
    },
    'xaxis': {
        'showgrid': False,
        'showline': False,
        'color': '#a6a6a6',
        'dividercolor': '#EBECF0'
    },
    'yaxis': {
        'showgrid': True,
        'tickformat': ',.0f',
        'gridcolor': '#EBECF0',
        'showline': False,
        'color': '#a6a6a6'
    },
    'separators': ',\''
}

figdict = {
    'data': [production_therm_and_renewable, production_nuclear, production_hydro, consumption, consumption_point],
    'layout': layout
}



fig = go.Figure(**figdict)
fig.add_annotation(
    dict(
        font=dict(color="black",size=14),
        x=1,
        y=df.consumption_country.iloc[-1],
        textangle=0,
        xanchor='left',
        showarrow=False,
        text='<b>Total consumption</b>',
        yshift=2,
        xref='paper'
    )
)

fig.show()
```
